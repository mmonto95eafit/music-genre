# Matriz de covarianzas y Outliers 

```python
x = df.loc[:, 'chroma_stft_mean':'mfcc20_var']
y  = df['label'].cat.codes.values
```

```python
def mahalanobis(x=None, data=None, cov=None):

    x_mu = x - np.mean(data)
    #if not cov:
        #cov = np.cov(x.T)
    inv_cov = np.linalg.inv(cov)
    mahalanobis_distances = np.diag(np.dot((x_mu @ inv_cov), x_mu.T))
   
    return mahalanobis_distances
```

```python
data = df.iloc[:,2:-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
train_og = model.score(x_train, y_train)
test_og = model.score(x_test, y_test)

print(train_og)
print(test_og)
```
