### Matriz de Covarianza



```python
X = np.array(df_x)
X
```

```python
S = np.cov(X.T)
S
```

### Definimos las funciones para calcular la matriz de covarianzas corregida, según la documentación del ejercicio

```python
def cov_var_corregida(X):
  m = len(X)  
  X_centrada = X-X.mean(axis=0)
  return (1/(m-1))*X_centrada.T@X_centrada
```

### Se calcula la matriz de covarianza corregida


```python

S_corregida = cov_var_corregida(X)
S_corregida
```

### Varianza total

```python
# se calcula sumando la traza de la matriz de covarianza, en este caso usaremos la corregida.

T = np.trace(S_corregida)
T
```

### Varianza promedio

```python
S2 = np.var(X)
S2
```

### Varianza Generalizada

```python
# se calcula la varianza generalizada usando la matriz de covarianza corregida

VG = np.linalg.det(S_corregida)
VG
```

### Varianza Promedio


```python
# se definen m,n 

m,n = X.shape
m,n

# se calcula la varianza promedio

VP = VG**(1/n)
VP
```

### Dependencias Lineales

Diagonal de la matriz de covarianzas

```python
diag_cov = np.diag(S.round(2))
diag_cov
```

Diagonal de la matriz de precisión

```python
diag_cov_inv = np.diag(np.linalg.inv(S.round(2)))
diag_cov_inv
```

```python
R2 =  (1 - 1/(diag_cov*diag_cov_inv)).round(4)
cols = df_x.columns
R2
```
